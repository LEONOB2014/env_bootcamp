```python
import pysal
import numpy as np

f = pysal.open(pysal.examples.get_path("stl_hom.txt"))

y = np.array(f.by_col['HR8893'])

w = pysal.open(pysal.examples.get_path("stl.gal")).read()

print ("\n\n")

print ("w = " + str(w) + "\n")  # print Weight object reference

mi = pysal.Moran(y, w, two_tailed=False)

print ("%.3f"%mi.I + "\n") # value of Moran's I

print (str(mi.EI) + "\n") # expected value under normality assumption

print ("%.5f"%mi.p_norm + "\n") # p-value of I under normality assumption

print (help(mi))
```
